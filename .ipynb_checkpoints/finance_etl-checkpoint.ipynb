```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
ticker = "Data/companies_fundamentals_data.csv"
ticker_df = pd.read_csv(ticker)
ticker_df.head()
```

```python
company = "Data/companylist.csv"
company_df = pd.read_csv(company)
company_df.head()
```

```python
# Create a filtered dataframe from specific columns
ticker_columns = ["ticker", "eps", "pe", "divyield", "ebitda"]
ticker_transform = ticker_df[ticker_columns].copy()

ticker_transform.set_index("ticker", inplace=True)

ticker_transform.head()
```

```python
company_columns = ["Symbol", "Name"]
company_transform = company_df[company_columns].copy()

# Set index
company_transform = company_transform.rename(columns={"Symbol": "ticker"})
company_transform.set_index("ticker", inplace=True)

company_transform.head()
```

```python
conn_string = "postgres:postgres@localhost:5432/NYSE_db"
engine = create_engine(f'postgresql://{conn_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
ticker_transform.to_sql(name='ticker_symbol', con=engine, if_exists='append', index=True)
```

```python
company_transform.to_sql(name='company', con=engine, if_exists='append', index=True)
```

```python
pd.read_sql_query('select * from ticker_symbol', con=engine).head()
```

```python
pd.read_sql_query('select * from company', con=engine).head()
```
